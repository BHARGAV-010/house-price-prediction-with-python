**BOSTON HOUSE PRICE PREDICTION**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import sklearn.datasets
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

now we import a data in pandas data frame

```python
boston_data = pd.read_csv('/content/BostonHousing.csv' , header = 0)
```

```python
#let's print first 5 ROW of the dataset
boston_data.head()
```

```python
#let's find a size of the data
boston_data.shape
```

here a medv is represted a final value of the data in thousand doller

```python
#let's check a any missing value in that dataset in
boston_data.isnull().sum()
```

```python
# Let's remove  a misssing value
boston_data = boston_data.dropna(how='any')
```

```python
#let's find out a ststicsl measure for the data..
boston_data.describe()
```

```python
#let's seperated a labels and data
X = boston_data.drop(['medv'] , axis=1)
Y = boston_data['medv']
```

```python
print(X)
print(Y)
```

```python
# Let's find a various type of corelation between all feature ...
correlation = boston_data.corr()
```

```python
# here a two type of correlation are present
# nagative corelation
# positive corelation (postive impact of that price and any feature)
```

```python
# let's cunstruct a heapmap for that that correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation ,cbar=True , square=True , fmt='.1f' , annot=True  , cmap='Greens')
# here a fmt is represnted a float value after point
```

```python
model = linear_model.LinearRegression()
```

```python
# Let's train that data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
# Let's check a shape of that all perameters
print(X.shape , X_train.shape , X_test.shape)
```

```python
model.fit(X_train,Y_train)
```

```python
# Let's make a predictive system for that data
```

```python
#prediction for tranning data
trainning_data_prediction = model.predict(X_train)
```

```python
print(trainning_data_prediction)
```

**here we can find a error for the our prediction**

if that error come to 5 or more that means our  data can not work properly

```python
# R squared error
score_1 = metrics.r2_score(Y_train, trainning_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, trainning_data_prediction)

print("R squared error for tranning data is :-  ", score_1)
print('Mean Absolute Error for tranning data is :- ', score_2)
```

let's do a same thing for test data

```python
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_3 = metrics.r2_score(Y_test , test_data_prediction)

# Mean absolute Error
score_4 = metrics.mean_absolute_error(Y_test , test_data_prediction)

print("R squared error for testning data is :-  ", score_3)
print('Mean Absolute Error for testning data is :- ', score_4)
```

```python
#that all error output are between 0-3 that means our model can work properly
```
